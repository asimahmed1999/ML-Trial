```python
df = pd.concat([aapl, msft], axis =1)
```

```python
df.corr()
```

```python
import seaborn as sns
df.drop(["microsoft_Volume", "apple_Volume"], axis=1, inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(7,7))
sns.heatmap(df.corr(), annot=True, linewidths=.5, ax=ax)
```

```python
sns.pairplot(data=df)
```

```python
aapl.apple_Open.plot(label='apple_Open')
msft.microsoft_Open.plot(label='microsoft_Open')
plt.legend()
```

```python
for i,j in zip(aapl.columns[1:],msft.columns[1:]):
    aapl[i].plot(label=i)
    msft[j].plot(label=j)
    plt.legend()
    plt.pause(0.01)
```

### Correlation 
Looking at the: 
1. subplots
2. heatmap
3. pairplots
4. side by side plot

We can see strong positive correlation between Microsoft(msft) and Apple Inc.(aapl)
